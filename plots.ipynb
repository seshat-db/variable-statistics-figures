```python

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.0f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct

def plot_pie(df, sum_row, highlight_row, labels, title, explode_val = False):
    palette = sns.color_palette("husl", 8)
    total = df.loc[sum_row].sum()
    part = df.loc[highlight_row].sum()
    data = [total - part, part]
    if explode_val:
        explode = [0, explode_val]
        plt.pie(data, labels=labels,
                colors=palette, explode=explode,
                autopct=make_autopct(data))
    else:
        plt.pie(data, labels=labels,
                colors=palette,
                autopct=make_autopct(data)) 
    plt.title(title)
    plt.show()
    
```

```python
matplotlib.use("pgf")
matplotlib.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
})

def save_plot_pie(df, sum_row, highlight_row, labels, title, explode_val = False):
    palette = sns.color_palette("husl", 8)
    total = df.loc[sum_row].sum()
    part = df.loc[highlight_row].sum()
    data = [total - part, part]
    if explode_val:
        explode = [0, explode_val]
        plt.pie(data, labels=labels,
                colors=palette, explode=explode,
                autopct=make_autopct(data))
    else:
        plt.pie(data, labels=labels,
                colors=palette,
                autopct=make_autopct(data)) 
    plt.savefig((title +'.pgf'))
```

```python
plt.clf()
save_plot_pie(info, "datapoints", "has_zotero", ("Unmatched", "Matched"), "DB citations with a match in the Zotero library", 0.03)
plt.clf()
save_plot_pie(info, "has_zotero", "has_PDF", ("has no PDF", "has PDF"), "Matched citations with a PDF in the Zotero library", 0.15)
plt.clf()
save_plot_pie(info, "datapoints", "has_pages", ("No Page numbers found", "Page numbers found"), "Citations with (found) page numbers", 0.05)
```
